```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Descarga de datos
tickers = ["AAPL", "MSFT", "GOOGL"] # Ejemplo de activos 
data = yf.download(tickers, start="2023-01-01", end="2025-12-31")
if 'Adj Close' in data.columns:
    df = data['Adj Close']
else:
    df = data['Close']
```

```python

# 3. Retornos y Gráficos
returns = data.pct_change().dropna()
cum_returns = (1 + returns).cumprod()

# Gráfico Retornos Acumulados
cum_returns.plot(figsize=(10, 6), title="Retornos Acumulados")
plt.show()

# Distribución de Retornos
plt.figure(figsize=(10,6))
sns.histplot(returns, kde=True)
plt.title("Distribución de Retornos")
plt.show()

# 4. Reporte de Calidad de Datos
print("--- Reporte de Calidad ---")
quality_report = pd.DataFrame({
    'Nulos': data.isnull().sum(),
    'Tipo': data.dtypes,
    'Registros': len(data)
})
print(quality_report)
```
